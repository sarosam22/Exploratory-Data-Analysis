```python
cmap = sns.cubehelix_palette(as_cmap=True)

##filter records less than 99th percentile of density to increase contrast
df_density_subset = df_white_wine_quality[df_white_wine_quality['density'] <= df_white_wine_quality['density'].quantile(0.99)]

fig, ax=plt.subplots(figsize=(8,8))
colors = ax.scatter(x=df_density_subset['alcohol'], y=df_density_subset['quality'], 
            c=df_density_subset['density'], 
                 cmap=cmap, alpha=0.3)
plt.xlabel('alcohol')
plt.ylabel('quality')
s=fig.colorbar(colors)
s.set_label('Density')


fit = np.polyfit(df_density_subset['alcohol'],df_density_subset['quality'],1)
fit_fn = np.poly1d(fit) 
# fit_fn is now a function which takes in x and returns an estimate for y

ax.plot(df_density_subset['alcohol'],df_density_subset['quality'], df_density_subset['alcohol'],
        fit_fn(df_density_subset['alcohol']))

```
